```python
# Step 1: Install libraries (optional)
!pip install pandas matplotlib scikit-learn

# Step 2: Create dataset.csv
data = """Hours,Score
2.5,21
5.1,47
3.2,27
8.5,75
3.5,30
1.5,20
9.2,88
5.5,60
8.3,81
2.7,25
7.7,85
5.9,62
4.5,41
3.3,42
1.1,17
8.9,95
2.5,30
1.9,24
6.1,67
7.4,69"""

with open("dataset.csv", "w") as file:
    file.write(data)

print("âœ… dataset.csv created!")

# Step 3: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# Step 4: Load the dataset
df = pd.read_csv("dataset.csv")
print("ðŸ“Š First 5 rows of the dataset:")
print(df.head())

# Step 5: Visualize the data
plt.scatter(df['Hours'], df['Score'])
plt.title("Study Hours vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.show()

# Step 6: Prepare data for ML
X = df[['Hours']]
y = df['Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 7: Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 8: Make predictions
y_pred = model.predict(X_test)

# Step 9: Show actual vs predicted
print("\nðŸ“ˆ Actual vs Predicted:")
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(result)

# Step 10: Evaluate the model
print("\nâŒ Mean Absolute Error:", mean_absolute_error(y_test, y_pred))

# Step 11: Predict for custom input
hours = float(input("\nâŒ¨ï¸ Enter number of study hours: "))
predicted_score = model.predict([[hours]])
print(f"ðŸŽ¯ Predicted Score = {predicted_score[0]:.2f}")
```
